```python
def new_engine(dialect, server, database, user=None, password=None, integrated_security=True):
    if integrated_security:
        eng = f"{dialect}://{server}/{database}?trusted_connection=yes&driver=ODBC+Driver+17+for+SQL+Server"
    else:
        assert user, 'You must define a username'
        assert password, 'You must define a password'
    print(eng)
    return create_engine(eng)

engine = create_engine("mssql://Raqi/Migration?trusted_connection=yes&driver=ODBC+Driver+17+for+SQL+Server")
print(engine)
```

```python
def execute_query(query, engine):
    try:
        return pd.read_sql_query(query, engine)
    except pd.errors.DatabaseError as db_err:
        print(f"Database error: {db_err}")
        return None
    except Exception as e:
        print(f"An error occurred: {e}")
        return None
```

```python
from contextlib import contextmanager

@contextmanager
def get_connection(engine):
    connection = engine.connect()
    try:
        yield connection
    except Exception as e:
        print(f"Error during connection: {e}")
    finally:
        connection.close()

def execute_query(query):
    with get_connection(engine) as connection:
        try:
            return pd.read_sql_query(query, connection)
        except Exception as e:
            print(f"Error executing query: {e}")
            return None
```

# Quota Refugees for Resettlement to Sweden and Germany through UNHCR, 2018-2023

A comparison of the number of quota refugees applications to resettle in Sweden and Germany, including calculations of the percentage change per year for each nationality to show increases or decreases over time across EU countries. Visualizations in the form of line and bar charts illustrate the changes over time.

```python
def get_refugee_data(country, start_year=2018, end_year=2023):
    query = f"""
    WITH refugeechange AS (
        SELECT
            Year,
            Country_of_origin,
            refugees_under_unhcr_s_mandate,
            Country_of_asylum,
            LAG(refugees_under_unhcr_s_mandate) OVER (PARTITION BY Country_of_origin, Country_of_asylum ORDER BY Year) AS Previous_Year_Refugees
        FROM
            kvotflyktingar{country}
        WHERE
            Year BETWEEN {start_year} AND {end_year}
    )
    SELECT
        Year,
        Country_of_origin,
        refugees_under_unhcr_s_mandate AS Current_Refugees,
        Previous_Year_Refugees,
        CASE 
            WHEN Previous_Year_Refugees IS NULL THEN NULL
            ELSE ROUND(((refugees_under_unhcr_s_mandate - Previous_Year_Refugees) / CAST(Previous_Year_Refugees AS FLOAT)) * 100, 2)
        END AS Percent_Change
    FROM
        refugeechange
    WHERE
        Previous_Year_Refugees IS NOT NULL
    ORDER BY
        Country_of_origin, Year;
    """
    return execute_query(query, engine)
```

# Asylum seekers applying for asylum in Sweden and Germany upon Arrival, 2018-2023

A comparison of the number of asylum seekers applying for asylum in Sweden and Germany, similar to the analysis of quota refugees. Calculations include annual percentage changes by nationality, showing increases or decreases over time across EU countries.
```python
def get_asylum_data(country, start_year=2018, end_year=2023):
    query = f"""
    WITH AsylumChange AS (
        SELECT
            Year,
            Country_of_origin,
            asylum_seekers,
            Country_of_asylum,
            LAG(asylum_seekers) OVER (PARTITION BY Country_of_origin, Country_of_asylum ORDER BY Year) AS Previous_Year_AsylumSeekers
        FROM
            asylumseeker{country}
        WHERE
            Year BETWEEN {start_year} AND {end_year}
    )
    SELECT
        Year,
        Country_of_origin,
        asylum_seekers AS Current_AsylumSeekers,
        Previous_Year_AsylumSeekers,
        CASE 
            WHEN Previous_Year_AsylumSeekers IS NULL THEN NULL
            ELSE ROUND(((asylum_seekers - Previous_Year_AsylumSeekers) / CAST(Previous_Year_AsylumSeekers AS FLOAT)) * 100, 2)
        END AS Percent_Change
    FROM
        AsylumChange
    WHERE
        Previous_Year_AsylumSeekers IS NOT NULL
    ORDER BY
        Country_of_origin, Year;
    """
    return fetch_data(query)
```

# Analysis of Quota Refugees and Asylum Seekers

Insights into the differences between quota refugees and asylum seekers to gain a comprehensive understanding of migration trends in Sweden and Germany.

```python
query_se = '''
    SELECT 
        kvotflyktingarSE.Year,
        SUM(CASE WHEN kvotflyktingarSE.Country_of_asylum = 'Sweden' THEN kvotflyktingarSE.Refugees_under_UNHCR_s_mandate ELSE 0 END) AS Total_Quota_Refugees,
        SUM(CASE WHEN asylumseekerSE.Country_of_asylum = 'Sweden' THEN asylumseekerSE.Asylum_seekers ELSE 0 END) AS Total_Asylum_Seekers
    FROM 
        kvotflyktingarSE
        JOIN asylumseekerSE ON kvotflyktingarSE.Year = asylumseekerSE.Year
    GROUP BY 
        kvotflyktingarSE.Year
    ORDER BY 
        kvotflyktingarSE.Year;
'''

df_se = pd.read_sql(query_se, engine)
```

```python
query_ger = '''
    SELECT 
        kvotflyktingarGER.Year,
        SUM(CASE WHEN kvotflyktingarGER.Country_of_asylum = 'Germany' THEN kvotflyktingarGER.Refugees_under_UNHCR_s_mandate ELSE 0 END) AS Total_Quota_Refugees,
        SUM(CASE WHEN asylumseekerGER.Country_of_asylum = 'Germany' THEN asylumseekerGER.Asylum_seekers ELSE 0 END) AS Total_Asylum_Seekers
    FROM 
        kvotflyktingarGER
        JOIN asylumseekerGER ON kvotflyktingarGER.Year = asylumseekerGER.Year
    GROUP BY 
        kvotflyktingarGER.Year
    ORDER BY 
        kvotflyktingarGER.Year;
'''

df_ger = pd.read_sql(query_ger, engine)
```

```python
# Sweden
df_se['Total'] = df_se['Total_Quota_Refugees'] + df_se['Total_Asylum_Seekers']
df_se['Quota_Refugee_Percentage'] = (df_se['Total_Quota_Refugees'] / df_se['Total']) * 100
df_se['Asylum_Seeker_Percentage'] = (df_se['Total_Asylum_Seekers'] / df_se['Total']) * 100

# Germany
df_ger['Total'] = df_ger['Total_Quota_Refugees'] + df_ger['Total_Asylum_Seekers']
df_ger['Quota_Refugee_Percentage'] = (df_ger['Total_Quota_Refugees'] / df_ger['Total']) * 100
df_ger['Asylum_Seeker_Percentage'] = (df_ger['Total_Asylum_Seekers'] / df_ger['Total']) * 100
```
